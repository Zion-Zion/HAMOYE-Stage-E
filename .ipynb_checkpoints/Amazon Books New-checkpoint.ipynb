```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
# load EDA packaes
import numpy as np
import pandas as pd

from time import sleep 
from selenium.webdriver.chrome.options import Options

# load web scrapping packages
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
from time import sleep
from timeit import default_timer as timer
```

```python
# since the first page and last page has the same pattern url for all the years 
base_url = "https://www.amazon.com/gp/bestsellers/{}/books"

years = [str(i) for i in range(2009, 2020)] # list of years between 2009-2019
urls = [base_url.format(year) for year in years] + \
       [base_url.format(year) + f"/ref=zg_bsar_pg_2/ref=zg_bsar_pg_2?ie=UTF8&pg=2" for year in years]
```

```python
def get_dir(book,year): 
    '''to get the details of each book for each year''' 
    
    import numpy as np
    '''to get the name of price'''

    try:
        price = book.find('span',class_="_cDEzb_p13n-sc-price_3mJ9Z").text[1:]
    except Exception as e:
        price = None
    try:
        ranks = book.find('span', class_='zg-bdg-text').text[1:]
    except Exception as e:
        ranks = None
    try:
        title = book.find('div',class_="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y").text
    except Exception as e:
        title = None
    try:
        ratings= book.find('span',class_="a-icon-alt").text[:3] 
    except Exception as e:
        ratings = None
    try:
        no_of_reviews  = book.find('span',class_="a-size-small").text
    except Exception as e:
        no_of_reviews = None
    try:
        author = book.find('a',class_="a-size-small a-link-child").text
    except Exception as e:
        author = None
    try:
        cover_type = book.find('span',class_="a-size-small a-color-secondary a-text-normal").text
    except Exception as e:
        cover_type = None
    year = year
    return [price,ranks,title,no_of_reviews,ratings,author,cover_type, year]


years = [str(i) for i in range(2009, 2020)] * 2
```

```python
year = [(str(i),str(i)) for i in range(2009,2020)] #create list that contains a set of each year
years = [j for i in year for j in i] #get a list from the above line
```

```python
# function to know how long it'll take to scrape the data

def format_time(seconds):
    minutes = seconds // 60
    seconds = seconds % 60
    if minutes > 0:
        return f"{minutes} minutes, {seconds} seconds"
    else:
        return f"{seconds} seconds"
```

```python
# start time
start_time = timer()

all_year = [] #Empty list of the content on the first and second page for the all the years of interest
for url in urls:  # loop through the urls created in cell 2
    
    website = url 
    
    driver =  webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) #to load the selenium webdriver 
    
    driver.get(website)        # use selenium webdriver above to get the webpage
    
    sleep(30)                  #to make sure the website is fullly loaded before going to the next page
    
    the_soup = BeautifulSoup(driver.page_source, 'html.parser')           #to get the page content
    
    books = the_soup.find_all(id = 'gridItemRoot')                 #get every books on the page.
    
    all_year.append(books) #add the books to the the list above
    
    
    
    driver.quit()                      #to close the chrome windows tab after extracting the data
    
# end time and print
end_time = timer()
elapsed_time = int(end_time - start_time)
print(f"Elapsed time: {format_time(elapsed_time)}")
```

```python
year_index = list(enumerate(years))
data = []

# Loop over each year and book, and extract the details
for year, book_list in zip(year_index, all_year):
    for book_page in book_list:
        for book in book_page:
            book_data = get_dir(book, year[1])
            data.append(book_data)
```

```python
# Open file for writing
with open('Amazon_data.txt', 'w+') as f:
     
    # Write each element of the data list to file
    for item in data: 
        try:
            f.write('%s\n' % item)
        except Exception as e:
            f.write('nothing\n')
     
    print("File written successfully")

 
# Close the file
f.close()
```

#### Loading the data

```python
# Create a DataFrame from the collected data, with appropriate column names
freq_bought_books = pd.DataFrame(data, columns=[
    'price',
    'rank',
    'title',
    'number_of_reviews',
    'rating',
    'author',
    'cover_type',
    'year'
])

# Rearrange the columns
freq_bought_books = freq_bought_books.reindex(columns=['title', 'author', 'rating', 'rank', 'number_of_reviews', 'price', 'cover_type','year'])
freq_bought_books

# Save the DataFrame to a CSV file
freq_bought_books.to_csv('freq_bought_books_2009-2019.csv', index=False)
```

```python
# create a copy of the dataset first
freq_bought_books_copy = freq_bought_books.copy()
```

### Issue 1

Dealing with null values in the dataset.
- For title column, the rows with Nan should be dropped.
- For cover_type, the row would have bee dropped while dropping rows with Null title enteries.
- For genre column, the values should be replaced with "Unknown"
- For price column, the price would be checked automatically and added to the dataset.

#### Code

```python
freq_bought_books_copy.dropna(subset=['title'], inplace=True)
```

```python
len(null_price_rows)
```

```python
# manually checked prices
price_list = [13.99,19,118.67,11.99,20.69, np.nan,8.39,11.99,np.nan,8.39,11.99,np.nan,np.nan,np.nan,22.62,np.nan,8.49,
              16.19,14.75,np.nan,np.nan,15.80,17.18,np.nan,np.nan,17.08,17.45,20.69,np.nan,17.08,30.99,np.nan,17.08,17.08,
              np.nan,17.08,np.nan,6.99,17.08]

# Not all books actually have a price that can be inputted

# Replace the missing values in the "price" column with the values from the list
null_price_rows = freq_bought_books_copy[freq_bought_books_copy["price"].isnull()]

null_price_rows_in = null_price_rows.index.tolist()
freq_bought_books_copy.loc[null_price_rows_in,"price"]=price_list
```

```python
freq_bought_books_copy['genre'] = freq_bought_books_copy['genre'].fillna("Unknown")
```

```python
# Replace missing values in 'author' column with values from 'number_of_reviews' column
freq_bought_books_copy['author'] = freq_bought_books_copy['author'].fillna(freq_bought_books_copy['number_of_reviews'])

# Drop the number_of_review column
freq_bought_books_copy = freq_bought_books_copy.drop('number_of_reviews', axis=1)
```
